```python
import pickle
import pandas as pd
# Step 1: Load the new dataset
data = pd.read_csv(r"C:\Users\suchi\Desktop\ELEVATE LABS\Classification-logisticreg\newdata.csv")
data

```

```python
# Load the trained model
with open("logistic_model.pkl", "rb") as f:
    model = pickle.load(f)

# Load the saved scaler
with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)
```

```python
# Step 2: Scale the new data
scaled_data = scaler.transform(data)

# OPTIONAL: If your model used PCA, load and apply it
# with open("pca.pkl", "rb") as f:
#     pca = pickle.load(f)
# transformed_data = pca.transform(scaled_data)

# If no PCA, just use the scaled data
input_data = scaled_data  # or use 'transformed_data' if using PCA

# Step 3: Make predictions
predictions = model.predict(input_data)

print(predictions)
for i, pred in enumerate(predictions):
    label = "Malignant" if pred == 1 else "Benign"
    print(f"Sample {i + 1}: {label}")


```
